44.Create Age_Group for customers (Child/Teen/Adult/Senior).

```python
def categorize_age(age):
    if age < 13:
        return 'Child'
    elif 13 <= age < 20:
        return 'Teen'
    elif 20 <= age < 60:
        return 'Adult'
    else:
        return 'Senior'

df['Age_Group'] = df['Age'].apply(categorize_age)
display(df[['Age', 'Age_Group']])
```

45.Create Loyalty_Level based on total purchase.

```python
def loyalty_level(amount):
    if amount < 500:
        return 'Bronze'
    elif 500 <= amount < 1000:
        return 'Silver'
    elif 1000 <= amount < 2500:
        return 'Gold'
    else:
        return 'Platinum'
df['Loyalty_Level'] = df['Total_Purchase'].apply(loyalty_level)
display(df[['Total_Purchase', 'Loyalty_Level']])
```

46.Encode Gender numerically (Male=1, Female=2, Other=3).

```python
def gender_to_numeric(gender):
    if gender == 'Male':
        return 1
    elif gender == 'Female':
        return 2
    else:
        return 3
df['Gender_Numeric'] = df['Gender'].apply(gender_to_numeric)
display(df[['Gender', 'Gender_Numeric']])
```

47.Encode Account_Status numerically.

```python
def account_status_to_numeric(status):
    if status == 'Active':
        return 1
    elif status == 'Inactive':
        return 2
    elif status == 'Banned':
        return 3
    else:
        return None
df['Account_Status_Numeric'] = df['Account_Status'].apply(account_status_to_numeric)
display(df[['Account_Status', 'Account_Status_Numeric']])
```
